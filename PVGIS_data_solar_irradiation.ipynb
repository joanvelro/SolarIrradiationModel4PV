```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

__Legend__
* __$E_d$__: Average daily electricity production from the given system (kWh)
* __$E_m$__: Average monthly electricity production from tge given system (kWh)
* __$H_d$__: Average daily sum of global irradiation per square meter received by the modules of the given system ($kWh/m^2$)
* __$H_m$__: Average monthly of global irradiation per square meter received by the modules  of the given system ($kWh_m^2$)

```python
data = {'Ed':[2.94,3.95,4.56,4.63,4.71,5.06,5.31,5.16,4.71,4.12,3.28,2.88,4.28],
        'Em':[91.2,111,141,139,146,152,164,160,141,128,98.5,89.1,130],
        'Hd':[3.64,4.97,5.94,6.16,6.41,7.02,7.52,7.27,6.46,5.46,4.17,3.57,5.72],
        'Hm':[113,139,184,185,199,211,233,226,194,169,125,111,174]}
index = ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun','Jul','Aug','Sept','Oct','Nov','Dec','year_av']
df = pd.DataFrame(data,columns=['Ed','Em','Hd','Hm'],index=index)
```

#### Plot data

```python
title = 'PVGIS'
df2 = df.ix[0:len(df)-1,'Ed']
ax = df2.plot(title=title,fontsize=2)
ax.set_xlabel("Time (Months)",fontsize=12)
ax.set_ylabel("$E_d$ Daily Energy Production ($kWh$)",fontsize=12)

ax.set_xticks(np.arange(0,13))
ax.set_xticklabels(index[0:len(index)-1],fontsize=12)

min_val = min(df.ix[0:len(df)-1,'Ed'])
max_val = max(df.ix[0:len(df)-1,'Ed'])
ax.set_yticks(np.linspace(min_val,max_val,12))

yt = np.linspace(min_val,max_val,12)
ylabelticks = [round(n, 2) for n in yt]
ax.set_yticklabels(ylabelticks,fontsize=12)

ax.legend()

plt.show()

```

```python
fig, ax1 = plt.subplots()

ax1.plot(np.arange(1,13),df.ix[0:12,'Ed'],'r-')

ax1.legend(loc='upper left')
min_val = min(df.ix[0:len(df)-1,'Ed'])
max_val = max(df.ix[0:len(df)-1,'Ed'])
ax1.set_yticks(np.linspace(min_val,max_val,12))
ax1.set_ylabel("$E_d$ Daily Energy Production ($kWh$)",fontsize=12)
ax1.set_xlabel("Time (Months)",fontsize=12)

ax2 = ax1.twinx()

ax2.plot(np.arange(1,13),df.ix[0:12,'Hd'],'b-')

ax2.legend(loc='upper right')
min_val = min(df.ix[0:len(df)-1,'Hd'])
max_val = max(df.ix[0:len(df)-1,'Hd'])
ax2.set_yticks(np.linspace(min_val,max_val,12))
ax2.set_ylabel("$H_d$ Daily Solar Irradiation ($kWh/m^2$)",fontsize=12)
ax2.set_xticks(np.arange(1,13))
ax2.set_xticklabels(index[0:len(index)],fontsize=12)

plt.show()
```
